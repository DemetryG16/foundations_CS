## 1. Create a single dataframe with the concatenation of all input csv files, adding a column called `country`

`d = pd.read_csv(f, encoding="utf-8-sig", encoding_errors="backslashreplace")` 
* we need the encoding due to the different languages, I used AI to choose the best encoding and errors.
* backslashreplace shows all characters even if they fail.

`csv = pd.concat(lista, ignore_index= True)`
* concat concatnates all data frames in the list in one table.
* ignore index reorder the indecies to be 0,1,2,....,n-1,n.


```python
pattern = r'trendingYT\\(\w+)videos.csv\\(\w+)videos.csv'
path = 'trendingYT/*.csv/*.csv'
files = glob.glob(path)  

lista = []    
country = []
for f in files:
    d = pd.read_csv(f, encoding="utf-8-sig", encoding_errors="backslashreplace")
    match = re.search(pattern , f)
    let = match.group(1)
    country.append(let)
    d['country'] = let
    lista.append(d)

csv = pd.concat(lista, ignore_index= True)
# csv
```

## 14. Read all json files with the video categories
IMPORTANT NOTES:
* Each id in the items in json files is connected directly to the category_id column in csv files
* Data in json files are nested. Check the column `items` in json using `json['items'][0]['snippet']['assignable']`  (I got this info from AI, I copied the text from json file and asked AI)
* The `id` in json files were not int64 and can't be combined with the video categories without transforming them first.
* json files don't have all `id` numbers that exists in `category_id`. So, make sure that `video_id` columns are preserved while merging.

```python
json_de = pd.read_json('trendingYT/DE_category_id.json')
j_path = glob.glob('trendingYT/*_category_id.json')
j_pat = re.compile(r'^trendingYT\\([\w]+)_category_id.json')

json_list = []
for j in j_path:
    jj = pd.read_json(j)
    match = j_pat.search(j)
    jj['country'] = match.group(1)
    json_list.append(jj)
    
json = pd.concat(json_list)
json.sort_index(inplace= True, ignore_index=True)

json['id'] = [json['items'][n]['id'] for n in range(len(json))]
json['id'] = json['id'].astype('int64')
json['assignable'] = [json['items'][n]['snippet']['assignable'] for n in range(len(json))]
```

```python
json.sort_values(by= 'id')
```

```python
csv_json = pd.merge(json,csv, how='right', left_on= ['id','country'], right_on= ['category_id', 'country'] ) 
# I got the left_on, right_on usage for 2 columns from AI as I did not understand that from pandas website.
# csv_json.dropna(inplace= True) # There are nan values
```

# The difference between json and csv files: (AI)
| Feature     | CSV                                                           | JSON                                                 |
| ----------- | ------------------------------------------------------------- | ---------------------------------------------------- |
| Format      | Comma-separated values (plain text)                           | JavaScript Object Notation (structured text)         |
| Structure   | Tabular (rows √ó columns)                                      | Hierarchical / nested (objects, arrays)              |
| Data types  | Mostly strings/numbers; no native support for booleans/arrays | Supports strings, numbers, booleans, arrays, objects |
| File size   | Usually smaller                                               | Can be bigger due to extra formatting                |
| Readability | Easy in Excel / pandas                                        | Human-readable but more verbose                      |
| Use case    | Simple tables / spreadsheets                                  | Complex, nested data (APIs, configs)                 |
